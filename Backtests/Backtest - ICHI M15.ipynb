```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./DE30_EUR.2019.M15', index_col=None, header=None)
df.columns = ['timestamp', 'Complete', 'Open', 'High', 'Low', 'Close', 'Volume']
df = df[['timestamp','Open','High','Low', 'Close', 'Volume']]
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.set_index(['timestamp'])
df.tail()
```

```python
df_backtest = df['2019-08-06']
df_backtest
```

```python
def Kijun(high, low):
    period_high = pd.Series(high).rolling(window= 26).max()
    period_low = pd.Series(low).rolling(window= 26).min()
    return (period_high + period_low) /2
```

```python
def Tenkan(high, low):
    period_high = pd.Series(high).rolling(window= 9).max()
    period_low = pd.Series(low).rolling(window= 9).min()
    return (period_high + period_low) /2
```

```python
def SenkouA(high, low):
    kijun = Kijun(high, low)
    tenkan = Tenkan(high, low)
    return ((tenkan+ kijun) / 2).shift(26)
```

```python
def SenkouB(high, low):
    period_high = pd.Series(high).rolling(window= 52).max()
    period_low = pd.Series(low).rolling(window= 52).min()
    return ((period_high + period_low) / 2).shift(52)
```

```python
from backtesting import Strategy
from backtesting.lib import resample_apply


class Ichimoku(Strategy):
    
    def init(self):
        # M15
        self.kijun_m15 = self.I(Kijun, self.data.High, self.data.Low)
        self.tenkan_m15 = self.I(Tenkan, self.data.High, self.data.Low)
        self.senkou_a_m15 = self.I(SenkouA, self.data.High, self.data.Low)
        self.senkou_b_m15 = self.I(SenkouB, self.data.High, self.data.Low)
        
        #H1
        self.kijun_h1 = resample_apply('H', Kijun, self.data.High, self.data.Low) 
        self.tenkan_h1 = resample_apply('H', Tenkan, self.data.High, self.data.Low)
        self.senkou_a_h1 = resample_apply('H', SenkouA, self.data.High, self.data.Low)
        self.senkou_b_h1 = resample_apply('H', SenkouB, self.data.High, self.data.Low)
        
        #H4
        self.kijun_h4 = resample_apply('4H', Kijun, self.data.High, self.data.Low)
        self.tenkan_h4 = resample_apply('4H', Tenkan, self.data.High, self.data.Low)
        self.senkou_a_h4 = resample_apply('4H', SenkouA, self.data.High, self.data.Low)
        self.senkou_b_h4 = resample_apply('4H', SenkouB, self.data.High, self.data.Low)
    
    def next(self):
        if False:
            print('ok')
```

```python
from backtesting import Backtest

bt = Backtest(df_backtest, Ichimoku, cash=10000, commission=.00015)
bt.run()
```
